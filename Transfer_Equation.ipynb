**ПРАКТИЧЕСКИЕ ЗАДАНИЯ**

$u'_t + \lambda u'_x = 0$

$h = e^{-2}$

$\sigma = 0.5$

$\tau = \sigma \cdot \frac{h}{\lambda}$

```
import numpy as np
import matplotlib.pyplot as plt

cur = 0.5
h = 1e-2
tau = cur*h
XX = int(2/h)
TT = 101 

def shema(a, b, c, f):

  Y = np.zeros((TT, XX))
  X = np.zeros(XX)
  T = np.zeros(TT)

  for i in range(0, TT):
    Y[i][0] = 0
    T[i] = i*tau
  for i in range(0, XX):
    if ((i*h > 0.4) and (i*h < 0.6)):
      Y[0][i] = 1
    else: 
      Y[0][i] = 0
    X[i] = i*h

  for i in range(0, TT-1):
    for j in range(1, XX-1):
      Y[i+1][j] = -a*Y[i][j-1] - b*Y[i][j] - c*Y[i][j+1] - f*Y[i+1][j-1]
    Y[i+1][XX-1] = Y[i][XX-2]

  return X, Y

```

```
YY = np.zeros(XX)
for i in range(0, XX):
  if ((i*h > 0.9) and (i*h < 1.1)):
    YY[i] = 1
```

**Задание 5.**

Теперь сделаем гибрид схем.

Используем сеточно-характеристический критерий монотонности.

$min(u_0^n, u_1^n) \leq u_m^{n+1} \leq max(u_0^n, u_1^n)$

где $u_0^n$ и $u_1^n$ точки с предыдущего временного слоя слева и справа от характеристики.



```
cur = 0.5
h = 1e-2
tau = cur*h
XX = int(2/h)
TT = 101

Y = np.zeros((TT, XX))
X = np.zeros(XX)
T = np.zeros(TT)

for i in range(0, TT):
  Y[i][0] = 0
  T[i] = i*tau
for i in range(0, XX):
  if ((i*h > 0.4) and (i*h < 0.6)):
    Y[0][i] = 1
  else: 
    Y[0][i] = 0
  X[i] = i*h

for i in range(0, TT-1):
  for j in range(1, XX-1):
    res1 = 1*Y[i][j-1] + 1/3*Y[i][j] + 0*Y[i][j+1] - 1/3*Y[i+1][j-1]
    res2 = 0*Y[i][j-1] + 1*Y[i][j] - 1/5*Y[i][j+1] + 1/5*Y[i+1][j-1]

    if ((min(Y[i-1][j], Y[i][j]) <= res1) and (res1 <= max(Y[i-1][j], Y[i][j]))):
      Y[i+1][j] = res1
    else:
      Y[i+1][j] = res2

  Y[i+1][XX-1] = Y[i][XX-2]

plt.figure(figsize = (15, 10))
plt.plot(X, Y[100, :], label='numerical solution')
plt.plot(X, YY, label='analitical solution')
plt.xlabel('X', fontsize=20)
plt.legend(fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()

```
